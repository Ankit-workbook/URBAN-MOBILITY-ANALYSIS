## GRAPHS

```python


# 1. Booking Status by Vehicle Type
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Vehicle_Type", hue="Booking_Status", palette="Set2")
plt.title("Booking Status by Vehicle Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("booking_status_vehicle.png", dpi=300)
plt.show()

# 2. Distribution of Ride Distance with Booking Status
plt.figure(figsize=(10,6))
sns.histplot(data=df, x="Ride_Distance", hue="Booking_Status", kde=True, bins=30, palette="Set1")
plt.title("Ride Distance Distribution by Booking Status")
plt.tight_layout()
plt.savefig("ride_distance_status.png", dpi=300)
plt.show()

# 3. Booking Value by Payment Method
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Payment_Method", y="Booking_Value", hue="Booking_Status", palette="Set3")
plt.title("Booking Value by Payment Method & Status")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("booking_value_payment.png", dpi=300)
plt.show()

# 4. Driver vs Customer Ratings (scatter)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Driver_Ratings", y="Customer_Rating", hue="Booking_Status", palette="coolwarm", alpha=0.7)
plt.title("Driver vs Customer Ratings by Booking Status")
plt.tight_layout()
plt.savefig("ratings_comparison.png", dpi=300)
plt.show()
```

## 📊 Insights from Visualizations

### 1. Booking Status by Vehicle Type
**Plot:** `booking_status_vehicle.png`

- Success rides dominate for **Bike, Mini, Prime Sedan**.  
- **Cancellations** by Driver/Customer are relatively higher in **Prime SUV & eBike**.  
- Smaller, cheaper vehicles (like **Bike, Mini**) seem more reliable for completed rides.  

---

### 2. Ride Distance Distribution by Booking Status
**Plot:** `ride_distance_status.png`

- Successful rides cluster around **short to mid distances (10–40 km)**.  
- Cancelled rides often have **0 distance** (likely canceled before starting).  
- Very long trips (**70+ km**) are rare but usually **successful** when booked.  

---

### 3. Booking Value by Payment Method & Status
**Plot:** `booking_value_payment.png`

- **Cash payments** dominate **low-value rides**.  
- **Credit Card & UPI** payments are more common for **higher-value rides**.  
- Failed/cancelled rides are seen across all methods, but are **more frequent with cash** (possibly due to last-minute dropouts).  

---

### 4. Driver vs Customer Ratings by Booking Status
**Plot:** `ratings_comparison.png`

- Strong **positive correlation** between driver & customer ratings.  
- Successful rides usually have **ratings between 3.5 – 5.0**.  
- Cancelled rides often show **missing or low ratings**, indicating dissatisfaction.  


```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
df['DayOfWeek'] = pd.to_datetime(df['Date']).dt.day_name()
df['Month'] = pd.to_datetime(df['Date']).dt.month
df['Week'] = pd.to_datetime(df['Date']).dt.isocalendar().week

```

```python
df['Total_TAT'] = df['V_TAT'] + df['C_TAT']
df['Speed'] = df['Ride_Distance'] / df['Total_TAT']
```

```python
df['Is_Cancelled'] = df['Booking_Status'].apply(lambda x: 0 if x=='Success' else 1)
df['Customer_Bookings'] = df.groupby('Customer_ID')['Booking_ID'].transform('count')
df['Route'] = df['Pickup_Location'] + " -> " + df['Drop_Location']
df['Route_Count'] = df.groupby('Route')['Booking_ID'].transform('count')

```

```python
df['Booking_Value_per_km'] = df['Booking_Value'] / df['Ride_Distance']
df['High_Value_Ride'] = df['Booking_Value'].apply(lambda x: 1 if x > 500 else 0)
```

```python
df['Rating_Diff'] = df['Driver_Ratings'] - df['Customer_Rating']
df['High_Rating'] = df['Customer_Rating'].apply(lambda x: 1 if x >= 4.5 else 0)
```

```python
df = pd.get_dummies(df, columns=['Vehicle_Type','Payment_Method','DayOfWeek'], drop_first=True)
```

```python
# Cancellation rate overall
cancellation_rate = df['Is_Cancelled'].mean()
print(f"Overall Cancellation Rate: {cancellation_rate*100:.2f}%")

# Cancellation rate by vehicle type
cancellation_by_vehicle = df.groupby('Vehicle_Type_eBike')['Is_Cancelled'].mean().sort_values(ascending=False)
print(cancellation_by_vehicle)
```

```python
# Correlation between numerical features
corr = df[['Ride_Distance','Booking_Value','Driver_Ratings','Customer_Rating','Total_TAT','Booking_Value_per_km']].corr()
print(corr)

# Optionally plot as heatmap


plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.savefig("Heatmap.png", dpi=300)
plt.show()
```

```python
# Average ride distance by Hour
hourly_distance = df.groupby('Hour')['Ride_Distance'].mean()
print(hourly_distance)

# Average cancellations by Day of Week
daily_cancellation = df.groupby('DayOfWeek_Monday')['Is_Cancelled'].mean()
print(daily_cancellation)
```

```python
# Average Booking Value by Vehicle Type
avg_value_vehicle = df.groupby('Vehicle_Type_eBike')['Booking_Value'].mean()
print(avg_value_vehicle)

# Average Customer Rating by Payment Method
avg_rating_payment = df.groupby('Payment_Method_UPI')['Customer_Rating'].mean()
print(avg_rating_payment)
```
